```python
# Py list Speed test
import numpy as np
import time 

size = 1_000_000_0

l1 = list(range(size))
l2 = list(range(size))

start = time.time()
add = [x+y for x,y in zip(l1,l2)]
end = time.time()
print(add[0:10])

print(end - start)
```

```python
# numpy array speedtest
import numpy as np
import time 

size = 1_000_000_0

l1 = np.array(list(range(size)))
l2 = np.array(list(range(size)))

start = time.time()
add = l1 + l2
end = time.time()
print (add[0:10])

print(end - start)
```

```python
import numpy as np
import time 

size = 1_000_000_0   # 10 million

l1 = list(range(size))

start = time.time()
list_squares = [x ** 2 for x in l1]
end = time.time()

print(list_squares[0:10])
print(end - start)
```

```python
import numpy as np
import time 

size = 1_000_000_0   # 10 million

arr1 = np.array(range(size))

start = time.time()
numpy_squares = arr1 ** 2
end = time.time()

print(numpy_squares[0:10])
print(end - start)
```
